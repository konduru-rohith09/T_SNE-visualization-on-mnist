```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test)=mnist.load_data()
```

```
x_train_reshape=x_train.reshape(x_train.shape[0], -1)
x_test_reshape=x_test.reshape(x_test.shape[0], -1)
```

```
standalized_data=StandardScaler()
x_train_scaled=standalized_data.fit_transform(x_train_reshape)
x_test_scaled=standalized_data.transform(x_test_reshape)
```

```
print(x_train_scaled.shape)
print(x_test_scaled.shape)
```

```
data=x_train_scaled[0:1000, :]
labels=y_train[0:1000]
```

```
tsne=TSNE(n_components=2, random_state=0)
tsne_data=tsne.fit_transform(data)
```

```
plt.figure(figsize=(8,6))
scatter=plt.scatter(tsne_data[:,0], tsne_data[:,1], c=labels, cmap="tab10",alpha=0.7)
plt.colorbar(scatter, ticks=range(10))
plt.xlabel("dimension 1")
plt.ylabel("dimension 2")
plt.title("t-sne visualization of mnist ")
plt.show()
```
